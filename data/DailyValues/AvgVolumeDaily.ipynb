```python
from matplotlib import pyplot as plt
import csv
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("AAPLDaily.csv")
df['timeAsDate'] = df['time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
print(df['timeAsDate'].apply(lambda x: x.year).unique())
df.plot(x='timeAsDate',y='volume')
```
